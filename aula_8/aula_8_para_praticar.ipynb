Abra o arquivo "Cancer_Data.csv" conte quantas linhas o dataframe possui com a coluna diagnosis = B.

Subistitua a coluna diagnosis seguindo a regra:
  "B" para benign cancer
  "M" para malignant cancer

Filtre os ids dos "malignant cancer" com raio maior que 25

```python
import pandas as pd 

df = pd.read_csv("Cancer_Data.csv")
df.head(10)
```

```python
cancer_b = df[df['diagnosis'] == 'B'].shape[0]


print("O total de linhas com diagnosis = B:", cancer_b)
```

```python


df['diagnosis']= df['diagnosis'].str.replace("B", "benign cancer")

df['diagnosis']= df['diagnosis'].str.replace("M", "malignant cancer")

df


```

```python

df_filtrado = df[(df['diagnosis'] == 'malignant cancer') & (df['radius_mean'] > 25)]

df_filtrado = df_filtrado['id']

print(df_filtrado)


```
