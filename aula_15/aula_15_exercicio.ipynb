```python
# exemplo aula

realtor = pd.read_csv('/Users/larissaferreira/coder_house/aula_13/realtor-data.csv')

features = (realtor 
            .groupby(['city', 'state'])
            .agg({'price': ['count', 'max', 'min', 'mean', 'median']})
            .reset_index())

features.columns = ['city', 'state', 'price_count', 'price_max', 'price_min', 'price_mean', 'price_median']

features.head(2)
```

```python
# a - Maior tamanho de casa por cidade

features = (realtor
            .groupby(['city', ])
            .agg({'house_size': ['max']})
            .reset_index()
            )

features.columns = ['city',  'house_size']

features.head()
```

```python
# b - Menor tamanho de casa por cidade

features = (realtor
            .groupby(['city', ])
            .agg({'house_size': ['min']})
            .reset_index()
            )

features.columns = ['city',  'house_size']

features.head()
```

```python
# c - Tamanho de casa médio por cidade

features = (realtor
            .groupby(['city', ])
            .agg({'house_size': ['mean']})
            .reset_index()
            )

features.columns = ['city',  'house_size']

features.head()
```

```python
features = (
    realtor.groupby('city')['house_size']
    .apply(lambda x: x.isnull().mean() * 100)
    .reset_index()
    .rename(columns={'house_size': 'missing_percentage'})
)

features.head()
```
