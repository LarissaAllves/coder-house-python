Usando o arquivo "best-selling game consoles.xlsx" conte quantos consoles tem na base por companhias.

```python
import pandas as pd 

df = pd.read_excel("/Users/larissaferreira/coder_house/aula_8/best_selling.xlsx",
                    sheet_name="consoles")
df.head(3)
```

```python
consoles_by_company = df['Company'].value_counts()

consoles_by_company_df = consoles_by_company.reset_index()
consoles_by_company_df.columns = ['Company', 'Contagem']  #renomeia a coluna
consoles_by_company_df
```

Ordene o resultado anterior pela quantidade de consoles por companhia de forma decrescente.

Sobre o dataframe anterior, adicione a coluna "Market Value" presente na tabela do arquivo "company.txt".

```python
df_company = pd.read_csv("company.txt",  sep=',')
df_company.head(3)
```

```python

df_merge = pd.merge(
    consoles_by_company_df, 
    df_company, 
    on="Company",  
    how="left"     
)
df_merge
```

Qual o valor médio de Market Value por console?

```python
# print("Valor Médio: ", dfConcat["Market Value"].median(numeric_only=True))


value = df_merge["Market Value"] / df_merge["Contagem"]

print("Valor médio", round(value.mean()))

```
