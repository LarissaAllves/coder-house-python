NumPy 
Biblioteca de codigo aberto que fornece suporte para trabalhar com arrays e matrizes multidimensionais, bem como funcoes matemáticas para operaçoes com arrays

NumPy Arrays

Os arrays podem ter qualquer numero de dimensões, permitindo que sejam usados para armazenar dados em matrizes e tensores

```python
# Ex.:

import numpy as np

np.array([1,2,3,4,5])
```

NumPy - Ndarrays

Tensores são uma extensão de matrizes para mais de duas dimensões. Tensor pode ter qualquer número de dimensões



```python
# Ex.: matriz

import numpy as np

matriz = np.array([[1.0,2], [3,4]])
matriz

```

```python
# Ex.: tensor

import numpy as np

tensor = np.array([[[1,2], [3,4]], [[5,6], [7,8]]])
tensor
```

NumPy - Ndarrays

dtype - é o atributo de um array que retorna o tipo de dados dos elementos do array

shape é o atributo que fornece o tamanho do array ao longo de cada dimensão

```python
print(matriz.dtype)
print(tensor.dtype)

print(matriz.shape)
print(tensor.shape)
```

NumPy - Criação de Ndarrays

```python
ones = np.ones((3,3))
ones
```

```python
full = np.full((3,2),42)
full
```

```python
zeros = np.zeros((2,3))
zeros
```

NumPy - Valores aleatórios

As principais formas são:

numpy.random.rand: gera valores aleatórios no intervalo [0,1].

numpy.random.randint: gera valores inteiros aleatórios em um intervalo e forma especificada

```python
np.random.rand(2,3)
```

```python
np.random.randint(0,10, size=(2,3))
```

NumPy - Indexação

A indexação em um array NumPy é semelhante à indexação em uma lista ou vetor, com algumas diferenças


```python

a = np.array([[1,2,3], [4,5,6]])

print('Indexação 1:')
print(a[0:2, 1:3])

print('Indexação 2:')
print(a[:, 0:2])

print('Indexação 3:')
print(a[1, :])

```

NumPy - Operações

```python

a = np.array([[1,2,3], [4,5,6]])

print("SUM: ", a.sum()) #soma
print("MAX: ", a.max()) # maxima
print("MIN: ", a.min ()) # minima
print("MEDIA: ", a.mean()) # media
print("STD: ", a.std()) # desvio padrão
print("VAR: ", a.var()) #variancia





```

Pandas

É uma biblioteca de código aberto que é usada para análise e manipulação de dados. Ela fornece estruturas de dados eficientes e de alta performance, juntamente com ferramentas para ler e escrever dados em diferentes formatos, como csvm excel, sql, entre outros

Pandas e NumPy

A biblioteca Pandas é construída sobre a biblioteca NumPy, o que significa que as estruturas de dados do Pandas são criadas a partir de arrays NumPy. Isso permite que as operações do NumPy sejam aplicadas as estruturas de dados do Pandas.

Portanto, pandas é uma extensão da biblioteca NumPy que adiciona recursos adicionais para a análise e manipulação de dados

Pandas - Serie

A Series é uma estrutura de dados unidimensional que pode armazenar diferentes tipos de dados

A principal diferença entre uma Series e um array NumPy é que uma Series tem um índice associado a cada valor

```python
# ex:
import pandas as pd

arr = np.array(range(10,20))
print(arr)


pd.Series(arr)
```

Pandas - DataFrame

DataFrame é uma estrutura de dados bidimensional que pode armazenar dados de diferentes tipos em diferentes colunas.

Cada coluna ou linha pode ser relacionada a uam serie, portanto DataFrame é a junção de várias Series

```python

mat = np.array([[1,2,3], [4,5,6]])

df = pd.DataFrame(mat)
df
```

```python
df = pd.DataFrame(mat, columns=["col1", "col2", "col3"])
df
```

```python
import pandas as pd

# Criando um dicionário com dados
dados = {
    'Nome': ['Ana', 'Bruno', 'Carlos'],
    'Idade': [25, 30, 45],
    'Cidade': ['Rio de Janeiro', 'São Paulo', 'Belo Horizonte']
}

# Convertendo o dicionário em um DataFrame
df = pd.DataFrame(dados)

# Exibindo o DataFrame
print(df)
```
