Visualização de Dados

Bibliotecas

Matplotlib - é umas das bibliotecas mais populares. Pode ser usada tanto em ambientes interativos como em scripts Python, e é altamente integrável com outras bibliotecas de análise de dados como NumPy e Pandas. É possível personalizar cada aspecto do gráfico. desde ps rótulos dos eixos até a cor e o tamanho dos pontos.

Interfaces

Orientada ao estado: para chamadas de função, maior compatibilidade com o MATLAB

Orientada a objetos: permite um maior grau de controle sobre os gráficos porque os tratamos como objetos



```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python

x = [1,2,3,4,5]
y = [10,8,6,4,2]

plt.plot(x,y)

plt.show()
```

```python
#EXEMPLO - ORIENTADA AO ESTADO

plt.plot([1,2], [3,4])
```

Tipos de Gráficos

Linhas - Mostra a tendencia ao longo do tempo ou do eixo x

Barras - É utilizado para comparar dados em diferentes categorias

Pizza - Mostra a distribuição percentual de dados em uma única categoria

Dispersão - Mostra a relação entre duas variáveis numéricas

Histograma - Mostra a distribuição de frequência de dados em um intervalo

Area - Mostra a tendência e a magnitude da mudança em dados ao longo de tempo

```python
#linhas
x = [1,2,3,4,5]
y = [10,8,6,4,2]

fig, ax = plt.subplots()

ax.plot(x,y)
```

```python
#barras
x = ["A", "B", "C","D", "E"]
y = [3,7,1,8,4]

fig, ax = plt.subplots()
ax.bar(x,y)
```

```python
#Pizza

labels = ['Maça', 'Bananans', 'Laranjas', 'Morangos']
quantidades = [20, 25, 10, 30]

fig, ax = plt.subplots()

ax.pie(quantidades, labels=labels)
```

```python
#Dispersão

x = [1,2,3,4,5]
y = [10,8,6,4,2]

fig, ax = plt.subplots()

ax.scatter(x,y)
```

```python
#Histograma

dados = [1,3,3,4,4,4,5,5,6,7,8,9]

fig, ax = plt.subplots()

ax.hist(dados, bins=4)
```

```python
#area

x = [1,2,3,4,5]
y = [2,4,6,8,10]

fig, ax = plt.subplots()

ax.fill_between(x,y)
```

```python
#Legendas
x = [1,2,3,4,5]
y = [10,8,6,4,2]

fig, ax = plt.subplots()

ax.plot(x,y)

#Definindo o titulo e etiqueta dos eixos

ax.set_title("Gráfico de dispersão")
ax.set_xlabel("Eixo x")
ax.set_ylabel("Eixo y")
```

Configurações

 - Titulo do gráfico: é definido usando a função title()
 - Etiqueta dos eixos: é definido usando a função xlabel() e ylabel()
 - Limites dos eixos: é definido usando a função xlim() ylim()
 - Escala dos eixos: é definido usando a função  xscale() yscale()
 - Legenda: é definido usando a função legend()
 - Cores: é definido usando o argumento color ou c
 - Estilo de linha: é definido usando o argumento linestyle ou ls
 - Tamanho do marcador: é definido usando o argumento marketsize ou ms
 - Tipo de marcador: é definido usando o argumento marker
 - Transparencia: é definido usando o argumento aplha
 - Espessura da linha: é definido usando o argumento linewidth ou lw
 - Grades: é definido usando a função grid()

```python
#EXEMPLO DE CONFIGS

x = [1,2,3,4,5]
y = [10,8,6,4,2]

fig, ax = plt.subplots()
ax.plot(x, y,
        color= 'black',
        linestyle='--',
        linewidth=2,
        marker='o',
        markersize=8,
        markerfacecolor='red'
        )
ax.set_title('Titulo do Grafico')
ax.set_xlabel('Eixo x')
ax.set_ylabel('Eixo y')
ax.set_xlim(0,6)
ax.set_ylim(0,12)
ax.grid(False)
```

Subplots

O método subplots() é usado para criar múltiplos gráficos em uma figura. Com o subplot(), podemos especificar o numero de linhas e colunas de gráficos e a posição de cada gráfico na figura.

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)


#Criação de subplots de linhas

fig, axs = plt.subplots(3,1, figsize=(8,10))
axs[0].plot(x, y1)
axs[0].set_title('Grafico de Seno')

axs[1].plot(x, y2)
axs[1].set_title('Grafico de Cosseno')

axs[2].plot(x, y3)
axs[2].set_title('Grafico de Tangente')
```

Seaborn

É uma biblioteca de visualização de dados baseada no Matplotlib, enquanto o matplotlib é uma biblioteca mais básica e flexível, o seaborn oferece uma interface mais simplificada e especializada para a criação de gráficos estatísticos atraentes e informativos.

Temas no seaborn

A função set_theme permite escolher entre diferentes temas pré-definidos para personalizar a aparência dos seus gráficos


Gráficos de Distribuição

Visualização de distribuição de dados: podemos usar o displot() para visualizar a distribuição de valores de uma variável

Visualização de relações entre variáveis: podemos usar o scatterplot() para visualizar a relação entre duas variáveis

Visualização de categorias: podemos usar o boxplot() para visualizar a distribuição de uma variável categórica.


```python
# temas - dark, white, darkgrid, whitegrid, ticks
import seaborn as sns

sns.set_theme(style='darkgrid')

x = np.random.normal(size=100)

#plotar histograma e distribuição

sns.displot(x, kde=True)
```

```python
#Visualização de relações entre variáveis

data = {'peso': [80,65,70,60,75,66,61,73,77],
        'altura': [180,170,175,167,172,169,166,178,169]}

df = pd.DataFrame(data)

#plotar scatter plot

sns.scatterplot(x='peso', y='altura', data=df)
```

```python

# Visualização de categorias: podemos usar o boxplot()

data = {'tipo': ['Venda', 'Venda', 'Compra', 'Compra', 'Venda', 'Compra'],
        'valor': [1000, 1500, 500, 800, 1200, 600]}

df = pd.DataFrame(data)

sns.boxplot(x='tipo', y='valor', data=df)
```
